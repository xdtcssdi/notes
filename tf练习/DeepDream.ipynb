# 1.2 图像编码处理

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
image_raw_data = tf.io.gfile.GFile('./image/gem.jpg','rb').read()

with tf.compat.v1.Session() as sess:
    img_data = tf.image.decode_jpeg(image_raw_data) # 解码jpg图片
    plt.imshow(img_data.eval()) # 显示rgb图像
    plt.show()
```

# 1.3图像缩放

```python
# method
# AREA = 'area' BICUBIC = 'bicubic' BILINEAR = 'bilinear' GAUSSIAN = 'gaussian'
#LANCZOS3 = 'lanczos3' LANCZOS5 = 'lanczos5' MITCHELLCUBIC = 'mitchellcubic' NEAREST_NEIGHBOR = 'nearest'
with tf.compat.v1.Session() as sess:
    resized1 = tf.image.resize(img_data, [256, 256])
    resized1 = np.array(resized1.eval(), dtype='uint8')
    plt.imshow(resized1)
    plt.show()
```

# 1.4 裁剪或填充后缩放
### 1.4.1 tf.image.resize_with_crop_or_pad(image,target_height, target_width)
### 1.4.2 目标尺寸小于原图像尺寸，则在中心裁剪

```python
with tf.compat.v1.Session() as sess:
    croped = tf.image.resize_with_crop_or_pad(img_data, 400, 400)
    plt.imshow(croped.eval())
    plt.show()
```

### 1.4.3 目标尺寸大于原图像尺寸，周围填充黑色

```python
with tf.compat.v1.Session() as sess:
    croped = tf.image.resize_with_crop_or_pad(img_data, 2000, 2000)
    plt.imshow(croped.eval())
    plt.show()
```

# 1.5 随机裁剪
### 1.5.1 tf.image.random_crop(image, size, seed=None, name=None)

```python
with tf.compat.v1.Session() as sess:
    random_croped1 = tf.image.random_crop(img_data, [600,600,3])
    plt.imshow(random_croped1.eval())
    plt.show()
```

# 1.6 随机翻转

### 1.6.1 左右翻转

```python
with tf.compat.v1.Session() as sess:
    plt.imshow(img_data.eval())
    plt.axis('off')
    plt.show()
    flip_left_right = tf.image.flip_left_right(img_data)
    plt.imshow(flip_left_right.eval())
    plt.axis('off')
    plt.show()
```

### 1.6.2 上下翻转

```python
with tf.compat.v1.Session() as sess:
    plt.imshow(img_data.eval())
    plt.axis('off')
    plt.show()
    flip_up_down = tf.image.flip_up_down(img_data)
    plt.imshow(flip_up_down.eval())
    plt.axis('off')
    plt.show()
```

# 1.7 改变对比度

```python
with tf.compat.v1.Session() as sess:
    plt.imshow(img_data.eval())
    plt.show()
    contrast = tf.image.adjust_contrast(img_data, 0.5)
    plt.imshow(contrast.eval())
    plt.show()
```
